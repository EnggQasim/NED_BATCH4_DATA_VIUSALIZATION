# Now Extract hmlt information from live websit

```python
import urllib.request   # not urllib - important!
url_text = urllib.request.urlopen('https://www.w3schools.com/')
print(type(url_text))
```

```python
from bs4 import BeautifulSoup
live_web = BeautifulSoup(url_text,'html.parser')
print(live_web.prettify())
```

```python
l1 = live_web.find_all("a")
print(len(l1))
```

```python
df_l = []
for tag in l1:
    df_l.append([tag.text,tag['href']])
    
df = pd.DataFrame(df_l, columns = ['text','links'])
print(len(df))
df
```

```python
df = df.drop_duplicates()
print(len(df))
```

```python
df.head()
```

```python
a = df.drop_duplicates("links")
print(len(a))
a.head()
```

```python
a = df.drop_duplicates("text")
print(len(a))
a.head()
```
