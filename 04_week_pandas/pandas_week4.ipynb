# apply custom function on subjects and return morethan one columns

````python
def my_grading_function(s1,s2,s3,s4,s5):
    total = 500 # var1
    obtained = s1 + s2 + s3 + s4 + s5 #var2
    per = obtained / total * 100 # var3
    
    grade = None # var4
    if per>=80:
        grade = "A+"
    elif per>=70:
        grade = "A"
    elif per>=60:
        grade = "B"
    elif per>=50:
        grade = "C"
    else:
        grade = "Fail"
        
    return obtained, total, per, grade

my_grading_function(88,95,87,90,78)
    
    
````
