# Easy to use

```python
a = [[1,2,3],
    [4,5,6],
    [7,8,9]]

display(a)
```

# How to create a basic array


```python
a = np.array([1,2,23,3,5])
print(type(a))
```

```python
a = np.zeros((10,10))
print(type(a))
a
```

```python
a = np.ones((10,10))
print(type(a))
a
```

```python
a = np.empty((10,10))
print(type(a))
a
```

```python
a = np.arange(1,11)
a
```

### Linspace

```python
a = np.linspace(1,2,5)
print(len(a))
a
```

```python
a = np.ones((2,3))
print(a.dtype)
a
```

```python
a = np.ones((2,3),dtype=np.int64)
print(a.dtype)
a
```
